## ReLU Activation Function

One of the most common [activation functions](https://www.codecademy.com/resources/docs/pytorch/nn/activation-functions) used in neural networks is called ReLU.

If a number is negative, ReLU returns `0`. If a number is positive, ReLU returns the number with no changes.

For example,

```
ReLU(-1)
# output: 0, since -1 is a negative number

ReLU(.5)
# output: .5, since .5 is not negative
```
